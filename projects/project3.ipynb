```python
tweets_list = []

miesiac = ['08', '09', '10', '11', '12']
miesiac2 = ['01', '02', '03', '04', '05']
dni = [8, 31, 32, 31, 32]
dni2 = [32, 29, 32, 31, 25]

for j in range(len(miesiac)):
    for d in range(1, dni[j]):
        if miesiac[j] == '08':
            a = 24+d
            q = 'Genshin Impact lang:en until:2022-{}-{}'.format(miesiac[j], str(a))
            for i,tweet in enumerate(sntwitter.TwitterSearchScraper(q).get_items()):
                if i > 149:
                    break
                tweets_list.append([tweet.date, tweet.id, tweet.content, tweet.user.username])
        else:
            if d <= 9:
                q = 'Genshin Impact lang:en until:2022-{}-0{}'.format(miesiac[j], d)
                for i,tweet in enumerate(sntwitter.TwitterSearchScraper(q).get_items()):
                    if i > 149:
                        break
                    tweets_list.append([tweet.date, tweet.id, tweet.content, tweet.user.username])
            else:
                q = 'Genshin Impact lang:en until:2022-{}-{}'.format(miesiac[j], d)
                for i,tweet in enumerate(sntwitter.TwitterSearchScraper(q).get_items()):
                    if i > 149:
                        break
                    tweets_list.append([tweet.date, tweet.id, tweet.content, tweet.user.username])
for j in range(len(miesiac2)):
    for d in range(1, dni2[j]):
        if d <= 9:
            q = 'Genshin Impact lang:en until:2023-{}-0{}'.format(miesiac2[j], d)
            for i,tweet in enumerate(sntwitter.TwitterSearchScraper(q).get_items()):
                if i > 149:
                    break
                tweets_list.append([tweet.date, tweet.id, tweet.content, tweet.user.username])
        else:
            q = 'Genshin Impact lang:en until:2023-{}-{}'.format(miesiac2[j], d)
            for i,tweet in enumerate(sntwitter.TwitterSearchScraper(q).get_items()):
                if i > 149:
                    break
                tweets_list.append([tweet.date, tweet.id, tweet.content, tweet.user.username])

tweets_df = pd.DataFrame(tweets_list, columns=['Datetime', 'Tweet Id', 'Text', 'Username'])
```

## Preprocessing
(tokenizacja, stopwords, lematyzacja)

```python
# lowercase
tweets_df["Text"] = tweets_df["Text"].apply(lambda x: x.lower())
```

```python
# tokenizacja
tweets_df["Tokens"] = tweets_df["Text"].apply(nltk.word_tokenize)
```

```python
# usunięcie linków
tweets_df["Tokens"] = tweets_df["Tokens"].apply(lambda x: [item for item in x if not '//' in item])
```

```python
# stopwords
stopwords = set(nltk.corpus.stopwords.words("english"))
additional = [".", ",", ":", ";", "'s", "``", "\''", "(", ")", "@", "[", "]", "#", 'https', "genshinimpact", "genshin", "impact", "honkaistarrail", "honkai", "star", "rail"]
stopwords.update(additional)
tweets_df["Tokens"] = tweets_df["Tokens"].apply(lambda x: [item for item in x if item not in stopwords])
```

```python
# lematyzacja
lemmatizer = WordNetLemmatizer()
tweets_df["Tokens"] = tweets_df["Tokens"].apply(lambda x: [lemmatizer.lemmatize(item, pos='v') for item in x])
```

```python
tweets_df["Tokens"] = tweets_df["Tokens"].apply(lambda words: ' '.join(words))
```

```python
tweets_df.to_csv("genshin.csv")
tweets_df
```

## Analiza opinii

### 1. Analiza całościowa

```python
sia = SentimentIntensityAnalyzer()
tweets_df["Sentiment"] = tweets_df["Tokens"].apply(lambda x: sia.polarity_scores(x)['compound'])
tweets_df['Sentiment']
```

```python
neg_df = tweets_df.query('Sentiment < 0.0')
pos_df = tweets_df.query('Sentiment > 0.0')
```

```python
text = " ".join(i for i in neg_df.Tokens)
wordcloud = WordCloud(stopwords=stopwords, width=800, height=400, background_color="white").generate(text)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title('Word Cloud from negative tweets')
plt.show()
# Można zauważyć mocne słowa (kill, hell, hate) i przekleństwa (shit, fuck, ass).
```

```python
text = " ".join(i for i in pos_df.Tokens)
wordcloud = WordCloud(stopwords=stopwords, width=800, height=400, background_color="white").generate(text)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title('Word Cloud from positive tweets')
plt.show()
# Pozytywne tweety wskazują na ogólną radość z gry (play, game, love, like, original). Widzimy słowa character, get, pull, które
# wskazują na zadowolenie z otrzymania jakiejś postaci.
```

```python
tweets_df["Emotion"] = tweets_df["Tokens"].apply(lambda x: te.get_emotion(x))
tweets_df["Happy"] = tweets_df["Emotion"].apply(lambda x: x['Happy'])
tweets_df["Angry"] = tweets_df["Emotion"].apply(lambda x: x['Angry'])
tweets_df["Surprise"] = tweets_df["Emotion"].apply(lambda x: x['Surprise'])
tweets_df["Sad"] = tweets_df["Emotion"].apply(lambda x: x['Sad'])
tweets_df["Fear"] = tweets_df["Emotion"].apply(lambda x: x['Fear'])
```

```python
maxindex = tweets_df[['Happy','Angry','Surprise','Sad','Fear']].idxmax(axis=1)
l = []
for i in tweets_df.index:
    l.append(maxindex[i])
tweets_df['Max'] = l
```
