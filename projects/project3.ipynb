## Preprocessing
(tokenizacja, stopwords, lematyzacja)

```python
# lowercase
tweets_df["Text"] = tweets_df["Text"].apply(lambda x: x.lower())
```

```python
# tokenizacja
tweets_df["Tokens"] = tweets_df["Text"].apply(nltk.word_tokenize)
```

```python
# usuniÄ™cie linkÃ³w
tweets_df["Tokens"] = tweets_df["Tokens"].apply(lambda x: [item for item in x if not '//' in item])
```

```python
# stopwords
stopwords = set(nltk.corpus.stopwords.words("english"))
additional = [".", ",", ":", ";", "'s", "``", "\''", "(", ")", "@", "[", "]", "#", 'https', "genshinimpact", "genshin", "impact", "honkaistarrail", "honkai", "star", "rail"]
stopwords.update(additional)
tweets_df["Tokens"] = tweets_df["Tokens"].apply(lambda x: [item for item in x if item not in stopwords])
```

```python
# lematyzacja
lemmatizer = WordNetLemmatizer()
tweets_df["Tokens"] = tweets_df["Tokens"].apply(lambda x: [lemmatizer.lemmatize(item, pos='v') for item in x])
```

```python
tweets_df["Tokens"] = tweets_df["Tokens"].apply(lambda words: ' '.join(words))
```

```python
tweets_df.to_csv("genshin.csv")
tweets_df
```

```python
tweets_df["Emotion"] = tweets_df["Tokens"].apply(lambda x: te.get_emotion(x))
tweets_df["Happy"] = tweets_df["Emotion"].apply(lambda x: x['Happy'])
tweets_df["Angry"] = tweets_df["Emotion"].apply(lambda x: x['Angry'])
tweets_df["Surprise"] = tweets_df["Emotion"].apply(lambda x: x['Surprise'])
tweets_df["Sad"] = tweets_df["Emotion"].apply(lambda x: x['Sad'])
tweets_df["Fear"] = tweets_df["Emotion"].apply(lambda x: x['Fear'])
```
