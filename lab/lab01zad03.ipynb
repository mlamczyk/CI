```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

## a)

```python
miasta = pd.read_csv('data/miasta.csv') # ładujemy dane z pliku csv i zapisujemy
print(miasta) # wyświetlamy tabelę
```

```python
print(miasta.values) # wyświetlamy czyste wartości tabeli
```

## b)

```python
miasta.loc['10', :] = [2010, 460, 555, 405] # dodajemy wiersz do tabeli
print(miasta)
```

## c)

```python
plt.plot(miasta['Rok'], miasta['Gdansk'], 'r.-') # wykres ludności od czasu w Gdańsku
plt.xlabel('Lata') # podspisujemy oś x
plt.ylabel('Liczba ludności [w tys.]') # podspisujemy oś y
plt.title('Ludność w Gdańsku') # dodajemy tytuł wykresu
plt.show() # wyświetlamy wykres
```

## d)

```python
plt.plot(miasta['Rok'], miasta['Gdansk'], 'r.-', label='Gdańsk') # etykiety będą widoczne w legendzie
plt.plot(miasta['Rok'], miasta['Poznan'], 'b.-', label='Poznań') # dodajemy wykres dla Poznania
plt.plot(miasta['Rok'], miasta['Szczecin'], 'y.-', label='Szczecin') # dodajemy wykres dla Szczecina
plt.xlabel('Lata')
plt.ylabel('Liczba ludności [w tys.]')
plt.title('Ludność w miastach Polski')
plt.legend() # widoczna legenda
plt.show()
```

## e)

```python
miasta.drop('Rok', axis=1, inplace=True) # usuwamy kolumnę z latami, chcemy skalować tylko liczbę ludności
std_scaler = StandardScaler() # skalowanie standardowe z bliblioteki sklearn
miasta1 = std_scaler.fit_transform(miasta.to_numpy()) # zamieniamy tabelę na tablicę ndarray i standaryzujemy
print('Średnia: ', round(np.mean(miasta1), 1)) # rozkład standardowy ma średnią równą 0
print('Odchylenie: ', np.std(miasta1)) # rozkład standardowy ma odchylenie równe 1
```

```python
minmax_scaler = MinMaxScaler() # skalowanie wartości od 0 do 1 z bliblioteki sklearn
miasta2 = minmax_scaler.fit_transform(miasta.to_numpy()) # normalizujemy
print('Minimum: ', np.min(miasta2)) # minimum powinno wynosić 0
print('Maksimum: ', np.max(miasta2)) # maksimum powinno wynosić 1
```
